## Punto 1.a

```python
r, theta = symbols('r theta')
m = symbols('m', integer=True, positive=True)
```

Radio

```python
rho = 7
```

Función de la condición de frontera

```python
f = cos(theta)**2
simplify(f)
```

Coeficiente $a_0$

```python
a0 = (1/pi) * integrate(f, (theta, -pi, pi))
simplify(a0)
```

Coeficientes $a_m$

```python
am = (1/(pi*(rho**m))) * integrate(f * cos(m*theta), (theta, -pi, pi))
simplify(am)
```

Coeficientes $b_m$

```python
bm = (1/(pi*(rho**m))) * integrate(f * sin(m*theta), (theta, -pi, pi))
simplify(bm)
```

Función $u(r,\theta)$

```python
u = (1/2) + ((r**2)/98) * cos(2*theta)
simplify(u)
```

Evaluación de la función sobre una grilla de puntos

```python
domain_radius = np.linspace(0, rho, 200)
domain_theta = np.linspace(-np.pi, np.pi, 200)
rd, th = np.meshgrid(domain_radius, domain_theta)
y = lambdify([r, theta], u, 'numpy')
```

Gráfico

```python
plt.subplots(figsize=(8, 8))
plt.subplot(projection="polar")
plt.pcolormesh(th, rd, y(rd,th), cmap=cm.coolwarm)
plt.grid()
plt.show()
```

## Punto 3.b

```python
np.set_printoptions(precision=4)
```

Función $f(x)$

```python
x = 0.5
f = lambda x: 2*x*sin(x) - cos(x)
```

Segunda derivada de $f(x)$

```python
fd2 = lambda x: 5*cos(x) - 2*x*sin(x)
fd2(x)
```

Convergencia 2da derivada con precisión 2

```python
H2 = np.array([])
EH2 = np.array([])
for j in range(1, 20+1):
    h = 2**(-j)
    f2 = (f(x-h) - 2*f(x) + f(x+h))/(h**2)
    Eh = np.abs(f2 - fd2(x))
    alpha = -1
    if j > 1:
        omega = Eh / EH2[j-2]
        phi = h / H2[j-2]
        alpha = np.log(float(omega)) / np.log(float(phi))
    print('{} {} {} {} {}'.format(j, h, f2, format(Eh, 'e'), alpha))
    H2 = np.append(H2, h)
    EH2 = np.append(EH2, Eh)
```

```python
plt.plot(H2, EH2)
plt.grid(True)
```

```python
plt.loglog(H2, EH2)
plt.grid(True)
```

Convergencia 2da derivada con precisión 4

```python
H4 = np.array([])
EH4 = np.array([])
for j in range(1, 20+1):
    h = 2**(-j)
    f2 = ((-1/12)*f(x-2*h) + (4/3)*f(x-h) - (5/2)*f(x) + (4/3)*f(x+h) - (1/12)*f(x+2*h))/(h**2)
    Eh = abs(f2 - fd2(x))
    alpha = -1
    if j > 1:
        omega = Eh / EH4[j-2]
        phi = h / H4[j-2]
        alpha = np.log(float(omega)) / np.log(float(phi))
    print('{} {} {} {} {}'.format(j, h, f2, format(Eh, 'e'), alpha))
    H4 = np.append(H4, h)
    EH4 = np.append(EH4, Eh)
```

```python
plt.plot(H4, EH4)
plt.grid(True)
```

```python
plt.loglog(H4, EH4)
plt.grid(True)
```

Convergencia 2da derivada con precisión 6

```python
H6 = np.array([])
EH6 = np.array([])
for j in range(1, 20+1):
    h = 2**(-j)
    f2 = ((1/90)*f(x-3*h) - (3/20)*f(x-2*h) + (3/2)*f(x-h) - (49/18)*f(x) + (3/2)*f(x+h) - (3/20)*f(x+2*h) + (1/90)*f(x+3*h))/(h**2)
    Eh = abs(f2 - fd2(x))
    alpha = -1
    if j > 1:
        omega = Eh / EH6[j-2]
        phi = h / H6[j-2]
        alpha = np.log(float(omega)) / np.log(float(phi))
    print('{} {} {} {} {}'.format(j, h, f2, format(Eh, 'e'), alpha))
    H6 = np.append(H6, h)
    EH6 = np.append(EH6, Eh)
```
